```python
print('EJERCICIO #11')

pal = str(input('escriba una frase: '))
print('\n')
import keyword #importa la libreria de palabras reservadas
import copy    #importar la libreria copy
 
v = ['A','E','I','O','U']   #lista de vocales mayusculas
vt = ['Á','É','Í','Ó','Ú']  #lista de vocales mayusculas con tilde
til = ['á','é','í','ó','ú'] #lista de vocales munusculas con tilde
kw = keyword.kwlist         #lista de palabras reservadas
nd = len(pal)               #longitud de la cadena   
copia = copy.deepcopy(pal)  #hace una copia profunda de la cadena ingresada
lcop = list(copia)          #hace una lista con las palabras de la copia de pal
ne = 0                      #contador de espacios                                    
nv = 0                      #contador de vocales mayusculas
nt = 0                      #contador de vocales con tilde 
nk = 0                      #contador de palabras reservadas

if type(pal) == str:        #confirma que la cadena ingresada es una cadena
    x = pal.split()         #hace una lista de palabras no toma espacios
    d = len(x)              #me da la cantidad de palabras en la lista
    ne = d-1                #la cantidad de palabras -1 me da los espacios
    for i in x:             #iterando atraves de la lista de solo palabras
        if i in kw:         #si el elemento esta en la lista de palabras reserv 
            nk +=1          #agregar uno al contador
for i in lcop:              #iterando atravez de la lista de la copia   
    if i in v:              #esta lista toma cada letra como elemento 
        nv +=1              #si el elemento esta en la lista de mayusculas
                            #agregar uno al contador       
    elif i in vt:           #si el elemento esta mayus con tilde 
        nv +=1              #agregar uno a vocales mayusculas
        nt +=1              #agregar uno a letras con tilde
        
    elif i in til:          #si esta en las minusculas con tilde
        nt +=1              #agregar uno al contador de tildes
if ne == -1:                #si el numero de espacios es -1 es decir,no hay 
    ne =0                   #asignar el valor de cero al numero de espacios  
#imprimir todos los contadores    
print('la longitud de la cadena es:',nd) 
print('el numero de vocales mayúsculas es',nv)
print('el numero de letras con tilde es:',nt) 
print('el numero de espacios entre palabras es',ne) 
print('el numero de palabras reservadas es:',nk)    
        
```

```python
#%%
import copy as cp

#%%
#funcion que imprime una matriz
def plot_Matriz(Matriz):
    '''
    la funcion solo reconoce matrices que ya esten definidas como matrices sin
    ningun tipo de error
    '''
    if type(Matriz) is list:
        for fila in Matriz:
            for columna in fila:
                print('{:8}'.format(columna),end = ' ')
#la funcion imprime en un formato de columnas luego de reconocer que el tipo de
#variable es una lista                
            print()
            
#esta funcion cuenta la cantidad de veces que esta un elemento en una lista            
def cont(L,num):
    '''
    L = una lista
    num = elemento para contar
    '''
    n = 0
    for i in L:
        if i ==num:
            n += 1
#la funcion cuenta la cantidad de veces que esta el numero en una lista y 
#retorna el contador            
    return n
#%%
print('EJERCICIO #18')
while True:
    try: #eval lee las entradas como una entrada valida de python
        A = eval(input('inserte una matriz cuadrada: '))
#sabemos que una matriz es una lista de listas        
        if type(A) is list: #si el tipo de variable es una lista
            for i in A:     #para cada elemento de la lista A
                if len(i)==len(A) and type(i) is list:
#si el tipo de elemento que esta en la lista A tambien es una lista
#si la longitud de los elementos de A y la longitud de la lista A se puede
#decir que la matriz es una matriz cuadrada                     
                    for j in i:
                        if type(j) is not int:
#para cada elemento de los elementos de A verificar que el tipo de variable
#sean un numero entero                            
                            raise Exception
                else: raise Exception
        else: raise Exception
#raise Exception para en caso de haber un error mandarlo a la entrada except
        break
#brack rompe el ciclo en caso de que no exista ningun tipo de error
    except:
        print('la matriz CUADRADA es una lista cuyas entradas sean listas\
              \ny las entradas de las listas sean numeros enteros')
U = cp.deepcopy(A)      
#crea una copia profunda de la matriz A 
print('la matriz ingresada es: ')
plot_Matriz(A) #llamamos la funcnion para que imprima A

for i in range(0,len(U)):            #en el rango de la longitud de U #elements
    for j in range(0,len(U[0])):     #y en el rango del primer elemento de U 
        if i == j:                   #si la posicion es la misma
            break                    #romper el ciclo 
        else:
            U[i][j] = 0              #sino cambiar el elemento por 0
print('la matriz U asociada es: ')
plot_Matriz(U) #imprime la matriz U que es a la cual se le hicieron los cambios    
n_1 =[] # lista con los contadores de la cantidad de unos en las filas
for i in U:            #para cada elemento de U
    n1 = cont(i,1)     #contar la cantidad de unos en cada elemento
    n_1.append(n1)     #agregarlos a la lista de contadores
    
maximo = 0             #variable del digito con el maximo de numeros 1
for maxi in n_1:       #para el valor maximo en la lista de contadores 
    if maxi>maximo:    #si el numero maxi es mayor que la variable maximo
        maximo = maxi  #la variable maximo toma el valor de maxi
        
d = n_1.index(maximo)  #arroja la posicion donde esta la cantidad maxima de 1's 
                         
print('la fila con mayor cantida de unos en la matriz U es: ',U[d])
#imprime la linea de texto y la fila donde esta la posicion con la mayor 
#cantidad de unos    
```
